<h2>Different $\Delta t$ Values</h2>
<p>One question we might as is 'what is the effect of the time step size,$\Delta t$.  We will now run the same simulation, but we will use successively larger values for $\Delta t$.</p>

<h3>Analysis of Error associated with $\Delta t$</h3>
<p>First, let us zoom in on the late stage of this graph where in is obvious that there is some error for the simulations with a larger $\Delta t$.  We will also include the exact analytic solution on this graph.</p>
<p>In the statement below, we define a function for the exact result, and then use that function to populate the p_exact list with the exact values.</p>

```python
def p(t):
    return 100* exp(0.1 * t)
t_exact = [t for t in range(101)]
p_exact = [p(t) for t in range(101)]
```

```python
start = time.perf_counter()
tEM, pEM = Chapter2.grow(1)
end = time.perf_counter()
timeEM=end-start
print("        Euler Time: {:10f}".format(timeEM))

start = time.perf_counter()
tRK2, pRK2 = Chapter2RK.grow(1)
end = time.perf_counter()
timeRK2=end-start
print("Runge - Kutta Time: {:10f}".format(timeRK2))
print("{:%}".format((timeEM-timeRK2)/timeEM))
```

```python
figure(figsize=(fsx, fsy))
plot(t_exact, p_exact, 'k-', label = "Exact")
plot(trk2, prk2, 'H', MFC=(1,0,0,0),mec='b', label='Runge - Kutta 2')
plot(tem, pem, 'H', MFC=(1,0,0,0),mec='r',label='Euler')
axis([80, 100, 000, 2500000])
xlabel('Time')
ylabel('Population')
title('Population vs Time')
legend(loc=2)
show()
```

```python
N=len(pem)
pem[0], prk2[0] = (100, 100)
errorEM  = [p(tEM[i])  - pEM[i]   for i in range(N)]
errorRK2 = [p(tRK2[i]) - pRK2[i]  for i in range(N)]
#errorc = [pc[i] - p(tc[i]) for i in range(N)]
#errord = [pd[i] - p(td[i]) for i in range(N)]
```

```python
figure(figsize=(fsx, fsy))
plot(tEM, errorEM, 'H', MFC=(1,0,0,0),mec='r', label='Euler')
plot(tRK2, errorRK2, 'H', MFC=(1,0,0,0),mec='b', label='Runge - Kutta 2')
#axis([80, 100, 000, 2500000])
xlabel('Time')
ylabel('Numerical Error')
title('Numerical Error vs Time')
legend(loc=2)
show()
```

<h3>Population at $t=100$ vs $\Delta t$</h3>
<p>To really understand the effect of the time step size, $\Delta t$ on the accuracy of the simulation, let us plot the simulation's prediction for the population at $t=100$ as a function of the time step size, $\Delta t$ </p>
<p>In the code below, the pop100 function is a custom function which returns only the popultation at $t = 100$.  Thus, each call to pop100 runs the simulation and returns the population for $t=100$.<p>

```python
print(inspect.getsource(Chapter2RK.pop100))
```

```python
p100exact=p(100)
steps = 2500
end = 5
delta = end/steps
dts=linspace(delta, end, steps)
RelErrorEM = [abs(Chapter2.pop100(dt)-p100exact)/p100exact for dt in dts]
RelErrorRK2 = [abs(Chapter2RK.pop100(dt)-p100exact)/p100exact for dt in dts]
```

```python
figure(figsize=(fsx, fsy))
plot(dts, RelErrorEM, 'H', MFC=(1,0,0,0),mec='r', label='Euler')
plot(dts, RelErrorRK2, 'H', MFC=(1,0,0,0),mec='b', label='Runge - Kutta 2')
xlabel('Delta t')
ylabel('Relative Error (Fraction)')
title('Relative Error at t = 100 vs Delta t')
legend(loc=2)
show()
```

<p>From this, we can see that simulation error from $\Delta t$ decreases as $\Delta t$ gets smaller.

```python
p100exact=p(100)
steps = 2000
end = .05
delta = end/steps
dts=linspace(delta, end, steps)
RelErrorEM = [abs(Chapter2.pop100(dt)-p100exact)/p100exact for dt in dts]
RelErrorRK2 = [abs(Chapter2RK.pop100(dt)-p100exact)/p100exact for dt in dts]
```

```python
figure(figsize=(fsx, fsy))
plot(dts, RelErrorEM, 'H', MFC=(1,0,0,0),mec='r', label='Euler')
plot(dts, RelErrorRK2, 'H', MFC=(1,0,0,0),mec='b', label='Runge - Kutta 2')
xlabel('Delta t')
ylabel('Relative Error (Fraction)')
title('Relative Error at t = 100 vs Delta t')
legend(loc=2)
show()
```

<h3>Runge-Kutta 4</h3>
<p>Let us repeat this demonstration with Runge - Kutta 4 routines.</p>

```python
def p(t):
    return 100* exp(0.1 * t)
t_exact = [t for t in range(101)]
p_exact = [p(t) for t in range(101)]
```

```python
start = time.perf_counter()
tEM, pEM = Chapter2.grow(1)
end = time.perf_counter()
timeEM=end-start
print("          Euler Time: {:12f}".format(timeEM))

start = time.perf_counter()
tRK2, pRK2 = Chapter2RK.grow(1)
end = time.perf_counter()
timeRK2=end-start
print("Runge - Kutta 2 Time: {:12f}".format(timeRK2))
print("{:%}".format((timeRK2-timeEM)/timeEM))

start = time.perf_counter()
tRK4, pRK4 = Chapter2RK4.grow(1)
end = time.perf_counter()
timeRK4=end-start
print("Runge - Kutta 4 Time: {:12f}".format(timeRK4))
print("{:%}".format((timeRK4 - timeEM)/timeEM))
```

```python
figure(figsize=(fsx, fsy))
plot(t_exact, p_exact, 'k-', label = "Exact")
plot(tRK4, pRK4, 'H', MFC=(1,0,0,0),mec='m', label='Runge - Kutta 4')
plot(tRK2, pRK2, 'H', MFC=(1,0,0,0),mec='b', label='Runge - Kutta 2')
plot(tEM,  pEM,  'H', MFC=(1,0,0,0),mec='r',label='Euler')
axis([80, 100, 000, 2500000])
xlabel('Time')
ylabel('Population')
title('Population vs Time')
legend(loc=2)
show()
```
